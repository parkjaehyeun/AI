```python
import pandas as pd
import matplotlib.pyplot as plt
a_series = pd.Series(['A','B','C','D','E','F','G'])
b_series = pd.Series([130,250,190,300,210,220,170])
c_series = pd.Series([1.9,2.6,2.2,2.9,2.4,2.3,2.2])
d_series = pd.Series([16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2])
df = pd.DataFrame({'name':a_series,'horse power' : b_series, 'weight' : c_series, 'efficiency' : d_series})
print(df)
## 생성한 시리지들을 모아 하나의 데이터프레임으로 만들어준다
## 파이썬 딕셔너리 구조로 열의 이름을 키로 데이터를 값으로 입력한다.
```

```python
x = df['horse power']
x = x.values.reshape(-1,1) #reshape를 해서 x를 2차원 array 형태로 바꿔준다
y = df['efficiency'] 
plt.plot(x,y,'o')
#참고사이트 https://hleecaster.com/ml-linear-regression-example/
```

```python
line_fitter = linear_model.LinearRegression()  #LinearRegression 모델 생성
line_fitter.fit(x,y)   #그 안에 x,y 데이터를 fit 시킨다
```

```python
print("계수 : ",line_fitter.coef_) #계수는 coef_를 통해서 알아낸다
print("절편 : ",line_fitter.intercept_) #절편은 intercept_를 통해서 알아낸다
score = line_fitter.score(x,y) # 예측점수는 line_fitter에 score함수를 사용해서 알아낸다
print("예측 점수 : ",score)
plt.plot(x,y,'o')
plt.plot(x,line_fitter.predict(x))
plt.show()
```

```python
a_series = pd.Series(['A','B','C','D','E','F','G'])
b_series = pd.Series([130,250,190,300,210,220,170])
c_series = pd.Series([1.9,2.6,2.2,2.9,2.4,2.3,2.2])
d_series = pd.Series([16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2])
w_series = pd.Series([1900,2600,2200,2900,2400,2300,2100])
df = pd.DataFrame({'name':a_series,'horse power' : b_series, 'weight' : c_series, 'efficiency' : d_series, 'weight':w_series})
#총중량(weight)를추가한 df를 생성

x = df[['horse power','weight']] #마력과 총중량을 x에 둠
y = df['efficiency'] #연비를 y에 둠
plt.plot(x,y,'o')
```

```python
line_fitter = linear_model.LinearRegression() #LinearRegression 모델 생성
line_fitter.fit(x,y)  #그안에 data를 fit하여 모델을 완성한다
print("계수 : ",line_fitter.coef_) #coef_를 통하여 기울기 구함
print("절편 : ",line_fitter.intercept_) #intercept_를 통해서 절편 구함
score2 = line_fitter.score(x,y) # 예측점수는 line_fitter에 score함수를 사용해서 알아낸다
print("예측 점수 : ",score2)
```

```python
a1 = 270 #마력 270
a2 = 2500 #총중량 2500
y_predicted = line_fitter.predict([[a1,a2]]) #위에서 만든 모델에 a1,a2를 넣어줘서 예측 연비를 구한다
y_predicted = np.round(y_predicted,2)  #예측연비를 소수점 둘째자리까지만 사용하도록 한다
print(a1," 마력",a2,"kg 자동차의 예상 연비 : ",float(y_predicted), " km/l")
```

```python
car = df[['efficiency','horse power','weight']] #마력과 총중량, 연비 사이의 상관관계를 그리기 위해서 df에서 필요한것들을 추출한다.
sns.pairplot(car) #데이터의 상관관계를 파악하기위해 seaborn.pairplot을 사용함

plt.show()
```

```python
sns.set(rc = {'figure.figsize':(7,7)}) #상관행렬 가시
c_matrix = car.corr().round(2) #상관행렬 생성
sns.heatmap(data = c_matrix, annot= True) #그후 seaborn.heatmap을 사용하여 상관관계를 가시화
plt.show()
```

seaborn.pairplot을 보면 중량(weight)이 2800 이상에서 2000 이하로으로 중량값이 감소 할수록 efficiency의 값이 대체로 증가하는 경향이 있다는점을 알수있다. 마력또한 300에서 150이하로 줄어들수록 중량과 유사하게 efficiency의 값이 증가하는 경향을 보인다.
seaborn.heatmap을 보면 중량과 마력은 대체로 비례하는 경향성을 띄고 있고 그와 반대로 연비는 중량과 마력에 반비례하는 경향성을 띄고있음을 알수있다.
