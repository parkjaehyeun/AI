```python
#LAB 6-1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
#필요한 패키지 임포트
data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc+'nonlinear.csv')

plt.scatter(df['x'],df['y'])# 데이터를 그려본다
```

```python
#데이터를 3차 다항 회귀를 위해 변환한다
from sklearn.preprocessing import PolynomialFeatures
X = df['x'].to_numpy()
y= df['y'].to_numpy()
X=X.reshape(-1,1) #입력을 2차원 벡터가 되게 한다 
feature_cubic= PolynomialFeatures(degree =3)
X_3 = feature_cubic.fit_transform(X)

#선형 회귀 모델을 만들고, 회귀 함수를 그릴 때 사용할 독립변수 범위를 0에서 1까지로 한다.
#이때 입력은 2차원 배열이 되어야 하므로 reshape(-1,1)을 지정한다.

lin_model = LinearRegression()
domain = np.linspace(0,1,100).reshape(-1,1)

#3차 다항 회귀를 위한 입력과 출력 레이블의 쌍으로 학습 실행, 회귀곡선 그림
lin_model.fit(X_3,y)
domain_3 = feature_cubic.fit_transform(domain)
predictions = lin_model.predict(domain_3)
plt.scatter(df['x'],df['y'])
plt.scatter(domain,predictions,color = )
```

```python
#LAB 6-4 비선형 SVM을 이용한 데이터 분류

#데이터 읽어오기
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'twisted_data.csv')
print(df.tail(5))
#입력데이터가 두개의 특징 x1,x2를 가지고
#각 데이터 인스턴스의 그룹은 y에 지정된 바와 같이 0또는 1임을 알수있다.

#데이터가 2차원 입력공간에 어떻게 섞여있는지 확인하기 위해 y=1인 그룹은 불긍ㄴ색 y=0인 그룹은 녹색으로 그려봄
df_positive = df[df['y']>0]
df_negative = df[df['y']==0]

import matplotlib.pyplot as plt
plt.scatter(df_positive['x1'],df_positive['x2'],color='r')
plt.scatter(df_negative['x1'],df_negative['x2'],color='g')

```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from yellowbrick.contrib.classifier import DecisionViz
X= df[['x1','x2']].to_numpy()
y= df['y']

#파이프라인을 이용하여 데이터 표준화와 다항 특징 변환을 한 뒤에 SVM 분류기에 넘기는형태로 분류기 생성
polynomial_svm_clf = Pipeline([
                               ("scaler", StandardScaler()),  #입력 데이터 표준화
                               ("poly_features", PolynomialFeatures(degree=5)), 
                               ("svm_clf", LinearSVC(C=1, loss= "hinge"))
])
polynomial_svm_clf.fit(X,y)
viz = DecisionViz(polynomial_svm_clf, title = "polynomial feature SVM") #분리 경계 확인
viz.fit(X,y)
viz.draw(X,y)
```

```python
# LAB 6-5 커널 트릭을 이용한 비선형 SVM

#데이터 읽어오기
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'twisted_data.csv')

from sklearn.svm import SVC #커널 트릭을 사용할 수 있는 SVM 분류기
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from yellowbrick.contrib.classifier import DecisionViz

X = df[['x1','x2']].to_numpy()
y= df['y']

polynomial_svm_clf = Pipeline([
                               ("scaler", StandardScaler()),  #입력 데이터 표준화
                               ("svm_clf", SVC(C=0.01, kernel = 'poly', degree =4, coef0 =2)) #커널트릭이 가능한 SVC로 생성, loss 키워드 매개변수없음, 다항커널의 경우 poly를 kernel키워드에, 다항 커널의 차수 지정, coef0은 커널에 있는 독립항 r을 의미
])
polynomial_svm_clf.fit(X,y)
viz = DecisionViz(polynomial_svm_clf, title = "polynomial feature SVM") #분리 경계 확인
viz.fit(X,y)
viz.draw(X,y)
```

6-2.
과적합이란 학습에 사용된 데이터에 대해서는 좋은 성능을 보이지만 새로운 데이터에 대해 적용하면 성능이 잘나오지 않는, 일반화 능력이 떨어지는 모델이 만들어지는것을 말하는데 이럴 경우 특징을 줄이거나, 데이터를 늘리고, 모델을 단순하게 만들어야한다 따라서 정트리가 과적합 하는 모습을 보일 경우 단순하 하기 위해 max_depth를 줄인다

6-3.
과소적합이란 모델이 지나치게 단순하거나 예측을 제대로 할수없는 특징들만 제공된 경우에 발생하므로 입력 데이터의 특징을 늘려 새로 학습하거나, 학습 모델이 가진 차수등의 복잡도를 높이는 방법으로 문제를 해결해야한다. 따라서 결정트리가 과소적합 하는 모습을 보일경우 모델의 복잡도를 높이기 위해 max_depth를 증가시켜야한다

6-4.
하드마진은 어떠한 데이터도 마진 내에 들어오지 않을 경우 부르는 마진을 뜻하는데, 하드마진에 가깝게 동작하려면 마진 내에 데이터가 침범할 수 있는 정도를 최소한으로 줄여야한다. 따라서 하드마진에 가깝게 동작하려면 이 C라는 매개변수를 줄일수있을만큼 최소한으로 줄여야한다.

```python
# 심화문제 6.6 

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features. We could
# avoid this ugly slicing by using a two-dim dataset
Y = iris.target
def my_kernel(X, Y):
    M = np.array([[2, 0], [0, 1.0]])
    return np.dot(np.dot(X, M), Y.T)

my_svm = SVC(kernel=my_kernel)
my_svm.fit(X, Y)
```
